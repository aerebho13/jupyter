```python
opening_brackets = ["[","{","("]
valid_pairs = [
    ["(", ")"],
    ["[", "]"],
    ["{", "}"]]

def balance_check(myStr):
    mystack = []
    for i in myStr:
        if i in opening_brackets:
            mystack.append(i)
        elif i in valid_pairs:
            pos = valid_pairs.index(i)
            if ((len(mystack) > 0) and
                (opening_brackets[pos] == mystack[len(mystack)-1])):
                mystack.pop()
            else:
                return "Unbalanced"
    if len(mystack) == 0:
        return "Balanced"
    else:
        return "Unbalanced"
```
