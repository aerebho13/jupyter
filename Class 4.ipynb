````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def append(self, item):
        if not self.head:
            self.head = Node(item)
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = Node(item)
            
    def insert(self, index, item):
        new_node = Node(item)
        if not self.head:    
            self.head = new_node
        elif index == 0:
            new_node.next = self.head
            self.head = new_node
        else:
            counter = 0
            current = self.head
            prev = None
            while current.next and index != counter:
                prev = current
                current = current.next
                counter += 1
            if index == counter:
                prev.next = new_node
                new_node.next = current
            else:
                current.next = new_node
                
    def remove(self, value):
        current = self.head
        if current.data == value:
            self.head = current.next           
        else:
            prev = None
            while current.next and current.data != value:
                prev = current
                current = current.next
            if current.data == value:
                prev.next = current.next
            else:
                raise ValueError("%s is not in list." % value)
                
    def __str__(self):
        output_str = "["
        current = self.head
        output_str += "%s " % current.data
        current = current.next
        while current:
            output_str += ", %s" % current.data
            current = current.next
        output_str += "]"
        return output_str
    
    def __len__(self):
        current = self.head
        length = 0
        while current:
            current = current.next
            length += 1
        return length
        
    def count(self, value):
        prev = self.head
        if current.data == value:
            self.head = current.next
        count = 0
        while prev:
            prev = prev.next
            count += 1
        return count
````

````python
mylist = LinkedList()
for i in range(5):
    mylist.append(i)

current = mylist.head
for i in range(5):
    assert current.data == i
    current = current.next

mylist.insert(0, -1)

assert mylist.head.data == -1

print(mylist)
print(len(mylist))
print(count(mylist))
````
